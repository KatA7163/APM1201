```python
from time import time
import matplotlib.pyplot as plt

# Code fragment 3.10
def find_max(data):
    """Return the maximum element from a nonempty Python list."""
    biggest = data[0]  # The initial value to beat
    for val in data:   # For each value:
        if val > biggest:  # if it is greater than the best so far,
            biggest = val  # we have found a new best (so far)
    return biggest  # When loop ends, biggest is the max

# Function to measure time
def measure_time(func, data):
    start_time = time()  # record the starting time
    func(data)           
    end_time = time()    # record the ending time
    elapsed = (end_time - start_time) * 1000  # compute the elapsed time
    return elapsed

input_sizes = [10, 50, 100, 500, 1000, 5000, 10000, 15000, 20000, 25000, 30000, 50000, 100000, 500000, 1000000]
times = []

for size in input_sizes:
    data = list(range(size))
    elapsed_time = measure_time(find_max, data)
    times.append(elapsed_time)
    print(f"Input size: {size}, Running time: {elapsed_time:.6f} milliseconds")

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, times, label='find_max', marker='o')
plt.xlabel('Input size (n)')
plt.ylabel('Running time (milliseconds)') 
plt.title('Running Time of find_max Function')
plt.legend()
plt.grid(True)
plt.xscale('linear')
plt.yscale('linear')
plt.savefig('find_max_times_plot.png')  
plt.show()
```
